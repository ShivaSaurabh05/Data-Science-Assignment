Note: Here Exception class refers to the base class for all the exceptions.

* using the Exception class as the base class for our custom exception ensures that our exception class has all of the necessary methods and attributes to work with the existing exception handling mechanisms in Python, and helps to maintain consistency and standardization in the way that exceptions are defined and raised in your code.

* By inheriting from the Exception class, our custom exception class automatically inherits all of these standard methods and attributes, making it easier to define and raise exceptions in a consistent and standardized way. 

Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(Exception)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

* ZeroDivisionError: Raised when trying to divide a number by zero.
* OverflowError: Raised when the result of an arithmetic operation is too large to be represented in the available memory.
* UnderflowError: Raised when the result of an arithmetic operation is too small to be represented in the available memory.
* FloatingPointError: Raised when a floating-point operation fails to produce a valid result, such as when trying to take the square root of a negative number.
* TimeoutError: Raised when an arithmetic operation times out or takes too long to complete.

```python
# ZeroDivisionError
try:
    10/0
except ZeroDivisionError as e:
    print(e)
```

```python
# OverFlowError (type of FloatingPointError)
a = 1.0
b = 10**500
try:
    c = a / b
except OverflowError as e:
    print(e)
```

```python
# Another example
try:
    a = (10.2-10.00)
    if a <0.2:
        raise FloatingPointError
except FloatingPointError as e:
    print("FloatingPointError")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

* The LookupError class is a built-in Python exception class that serves as a base class for several specific exception classes that are raised when a lookup operation fails, such as when a key is not found in a dictionary or when an index is out of range in a sequence.

```python
# KeyError example
try:
    d = {1:2,'name':'shiva'}
    print(d[2])
except KeyError as e:
    print(e)
```

```python
# IndexError
l = [1,2,3]
try:
    print(l[3])
except IndexError as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

* ImportError is a built-in Python exception class that is raised when there is a problem importing a module in a Python program. This could be due to a variety of reasons, such as a missing module, a circular import, or a version mismatch between the module being imported and the Python interpreter.

* ModuleNotFoundError is a more specific type of ImportError that is raised when a module cannot be found, while ImportError is a more general exception that can be raised for any problem related to importing a module.

```python
try:
    import shivaSaurabh
except ModuleNotFoundError:
    print("Error: Module not found")
else:
    print("Module imported successfully")
```
